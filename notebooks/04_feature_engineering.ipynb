```python
import yfinance as yf
import pandas as pd

# Choose stock (change if you want)
TICKER = "AAPL"

# Download 6 months of daily data
df = yf.download(TICKER, period="6mo", interval="1d")

# Reset index so Date becomes a column
df.reset_index(inplace=True)

# Save historical prices
df.to_csv("../data/raw/prices/historical_prices.csv", index=False)

print("Historical prices saved:", df.shape)
df.head()
```

```python
import pandas as pd

df = pd.read_csv("../data/raw/prices/historical_prices.csv")

# Create target
df["target"] = (df["Close"].shift(-1) > df["Close"]).astype(int)
df.dropna(inplace=True)

# Select ONLY available columns
features = df[[
    "Open",
    "High",
    "Low",
    "Close",
    "Volume",
    "target"
]]

features.to_csv("../data/processed/features/features.csv", index=False)

print("✅ Features generated:", features.shape)
features.head()
```
